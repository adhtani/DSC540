### 5. Handling Missing Values

I created a function to check the percentage of missing values in each feature. deleing null values is always risky, it needs careful analysis. I attempt to check which feature has highest missing values an ddepending on it's important I will either delete or keep the features

```python
# check for missing value
def percentMissing(df):
    
    df_numeric = df.select_dtypes(include=[np.number])
    numeric_cols = df_numeric.columns.values
    print(numeric_cols)
    
    # % of missing data
    for col in df.columns:
        # create missing indicator for features with missing data
        missing = df[col].isnull()
        pct_missing = np.mean(missing)*100
        print('{} - {}%'.format(col, round(pct_missing)))
        num_missing = np.sum(missing)
        
    return df
```

### 6. Stripping string data of whitespace

```python
def no_whitespace(string):
    return string.strip()
```

```python
# stripping white space from all column type as object
def clean_string(df):
    
    for col in df.columns:
        if df[col].dtype == np.object:
            new = '{}_new'.format(col)
            df[new] = df[col].apply(no_whitespace)
            df[col] = df[new]
            df = df.drop(columns=[new])
    return df


df1 = clean_string(df1)
df1.head(2)
```

#### Token Sort Method

Part 1: Find 90% match using Token Sort method and replace values 

```python
from fuzzywuzzy import process, fuzz
```

```python
def token_high(df, percent):
    
    # create a subset dataframe with higher percent match values
    high_score = df[(df['score'] >= percent) &
                                  (df['industry'] !=  df['match']) &
                                  (df['sorted_industry'] != df['match'])]
    
    # Drop the representative value column
    high_score = high_score.drop('sorted_industry',axis=1).copy()
    
    return high_score
```

```python
def token_fuzzy(df, percent, scorer):
    
    # unique industry lists
    unique_industry = df['industry'].unique().tolist()
    sorted(unique_industry)
    
    '''Token Ratio'''
    #Create tuples of industries, matched industries, and the score of token sort
    score = [(x,) + i
             for x in unique_industry 
             for i in process.extract(x, unique_industry, scorer=scorer)]

    #Create a dataframe from the tuples
    similarity = pd.DataFrame(score, columns=['industry','match','score'])

    # because this will have repitative pairs, we will remove duplicated pairs from this
    similarity['sorted_industry'] = np.minimum(similarity['industry'], similarity['match'])
    
    # Token high score
    high_score = token_high(similarity, percent)
    
    
    return similarity, high_score
```

```python
similarity, high_score = token_fuzzy(df1, 90, fuzz.token_sort_ratio)
high_score.groupby(['industry','score']).agg({'match': ', '.join}).sort_values(['score'], ascending=False)
```
