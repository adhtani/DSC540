### 2. Combine 3 datasets in sqlite database

```python
# counting distinct values that got combined
x = pd.read_sql('''SELECT COUNT(DISTINCT SI.ticker) FROM Stock_Indicator SI JOIN Company_Info CI ON CI.ticker = SI.ticker JOIN Daily_Prices DP ON SI.ticker = DP.ticker''', sqlite_connection)
x
```

```python
# counting total values that got combined
y = pd.read_sql('''SELECT COUNT(*) FROM Stock_Indicator SI JOIN Company_Info CI ON CI.ticker = SI.ticker JOIN Daily_Prices DP ON SI.ticker = DP.ticker''', sqlite_connection)
y
```

 **Combining All 3 dataframes**

```python
merged_df = pd.read_sql('''SELECT * FROM Stock_Indicator SI JOIN Company_Info CI ON CI.ticker = SI.ticker JOIN Daily_Prices DP ON SI.ticker = DP.ticker''', sqlite_connection)
merged_df.head()
```

```python
merged_df.drop('index', axis=1, inplace=True)
merged_df.head(3)
```

### 3. Visualization of the Stock Data

**1. Barplot of Market Capitalization by Sector**

Sector field is from the Company_Info table and Mkt_Cap is from Stock_Indicator table

```python
sector = merged_df.groupby('sector').sum()['Mkt_Cap']
```

```python
sector = pd.DataFrame(sector)
sector.index
```

```python
# set figure size
plt.rcParams['figure.figsize'] = (20,10)
sns.color_palette("Blues", as_cmap=True)
sns.barplot(x = 'Mkt_Cap', y = sector.index, data = sector, palette=("Blues"))

plt.xlabel("Market Capitalization")
plt.ylabel("Sector")
plt.title("Market Capitalization by sector as of March 06 2021")
# Show the plot
plt.show()
```

**2. Line chart of Daily Volume of Trades**

Sector field is from the Company_Info table and volume is from Daily Prices table

```python
# Grouping by sector for mean volume trades
sector_volume = merged_df.groupby(['sector', 'date']).mean()
```

```python
sector_volume = pd.DataFrame(sector_volume)
sector_volume.index
```

```python
sector_volume = sector_volume.reset_index(level=['sector', 'date'])
```

```python
# set figure size
plt.rcParams['figure.figsize'] = (20, 10)

sns.lineplot(x=sector_volume['date'], y=sector_volume['volume'], hue=sector_volume['sector'], palette=("Paired"))
plt.title("Daily Volume for Each Sector over the period of 2 weeks")
plt.xlabel("Date")
plt.ylabel("Volume of Daily Trades")
plt.show()
```

**3. Pie Chart of Median Employee in Each Sector**

Both field sector and num_employee is from Company_Info table

```python
sector_emp = merged_df.groupby(['sector']).median()['num_employees']
```

```python
sector_emp = pd.DataFrame(sector_emp)
sector_emp = sector_emp.reset_index(level=['sector'])
sector_emp
```

```python
# set figure size
plt.rcParams['figure.figsize'] = (10, 10)

plt.pie(sector_emp['num_employees'], labels=sector_emp['sector'], autopct='%1.1f%%',
        startangle=90)
plt.title('Median Number of Employees percentage in each sector')
plt.show()
```

**4. Histogram for Beta**

```python
df = pd.read_sql('''SELECT * FROM Stock_Indicator SI''', sqlite_connection)
df.head()
```

```python
# set figure size
plt.rcParams['figure.figsize'] = (20, 10)

sns.histplot(data=df, x="Beta_1Yr", kde=True)
plt.title("Beta Distribution for all Companies over 2B Market Cap")
plt.xlabel("1 Year Beta")
plt.ylabel("Count")
plt.show()
```

**5. Scatter Plot for vwap vs. daily change**

```python
df1 = pd.read_sql('''SELECT * FROM Daily_Prices DP''', sqlite_connection)
df1.head()
```

```python
ticker_prices = df1.groupby(['ticker']).mean()
```

```python
ticker_prices.head(2)
```

```python
# set figure size
plt.rcParams['figure.figsize'] = (20, 20)
sns.relplot(x="change", y="vwap", data=ticker_prices);

plt.title("Volume-weighted Average Price vs. Daily Change")
plt.xlabel("Daily Change")
plt.ylabel("Volume-weighted Average Price")
plt.show()
```

For merging the data, I had to do inner join because all three of my datasets had slightly different number of ticker symbols. The flat .csv file had 1388 ticker symbols, while html data had 2225 ticker symbols and API data had 2095 ticker symbols. This was due to API returning errors for missing symbols, and there is always possibility of companies getting delisted, in which case those symbols will not have any data. Due to this, I decided to do inner join so it will only keep the values that have ticker symbols in all 3 datasets. The combined dataset had distinct 1225 ticker symbols and 11024 records.
